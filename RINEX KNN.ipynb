```python
#declare feature vector and target variable
x=df
y=df['status_type']
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convert the DataFrame to a NumPy array
x_array = np.array(x)

# Assuming the kmeans model is already trained with k clusters
k = 3
labels = kmeans.labels_

plt.figure(figsize=(10, 5))

# Plotting the points for each cluster
for i in range(k):
    plt.scatter(x_array[labels == i, 0], x_array[labels == i, 1], label=f'Cluster {i}')

# Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')

plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering with 3 Clusters')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Assuming x is already preprocessed and scaled
x_array = np.array(x)

# Train the K-Means model with 4 clusters
k = 4
kmeans = KMeans(n_clusters=k, random_state=0)
kmeans.fit(x_array)

# Get the labels
labels = kmeans.labels_

# Create the visualization
plt.figure(figsize=(10, 5))

# Plotting the points for each cluster
for i in range(k):
    plt.scatter(x_array[labels == i, 0], x_array[labels == i, 1], label=f'Cluster {i}')

# Plotting the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')

plt.legend()
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering with 4 Clusters')
plt.show()
```
